```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z= np.linspace(-10, 10, 50)
gz= 1/(1+np.exp(-z))
```

```python
plt.plot(z, gz)
plt.title("Sigmoid Function", fontsize=14)
plt.xlabel("z", fontsize=14)
plt.ylabel("g(z)", fontsize=14)
plt.grid(True)
plt.show()
```

```python
x=np.array([0.053,0.081,0.096,0.158,0.242,0.253,0.396,0.427,0.459,0.572])
y= np.array([0,0,0,0,1,0,1,1,1,1])
```

```python

plt.figure(figsize=(5,5))
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)
plt.plot(x[y==0], y[y==0], "bs", markersize=6)
plt.plot(x[y==1], y[y==1], "g^", markersize=6)
plt.grid(True)
plt.show()  
```

```python
m= x.size
thetha1= np.arange(-40,60,0.15)
h= np.array([-0.5+thetha*x for thetha in thetha1])
j=(1/(2*m))*((h-y)**2).sum(axis=1)

plt.figure(figsize=(5,5))
plt.xlabel("Theta", fontsize=12)
plt.ylabel("Cost", fontsize=12)
plt.plot(thetha1, j, "b-", linewidth=2)
plt.grid(True)
plt.show()
```

```python
m= x.size
thetha1= np.arange(-40,60,0.15)
h= np.array([-0.5+thetha*x for thetha in thetha1])
h= 1/(1+np.exp(-h))
#MSE
#j=(1/(2*m))*((h-y)**2).sum(axis=1)
j= - (1/m)*(y*np.log(h) + (1-y)*np.log(1-h)).sum(axis=1)


plt.figure(figsize=(5,5))
plt.xlabel("Theta", fontsize=12)
plt.ylabel("Cost", fontsize=12)
plt.plot(thetha1, j, "b-", linewidth=2)
plt.grid(True)
```

```python
h= np.arange(0.01,1,0.01)
j=-np.log(h)
plt.figure(figsize=(5,5))
plt.xlabel("h", fontsize=14)
plt.ylabel("Cost", fontsize=14)
plt.title("y==1", fontsize=16)
plt.plot(h, j, "b-", linewidth=2)
plt.grid(True)
plt.show()
```

```python
h= np.arange(0.01,1,0.01)
j=-np.log(1-h)
plt.figure(figsize=(5,5))
plt.xlabel("h", fontsize=14)
plt.ylabel("Cost", fontsize=14)
plt.title("y==1", fontsize=16)
plt.plot(h, j, "b-", linewidth=2)
plt.grid(True)
plt.show()
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
thetha=np.array([4.3])

x_new=np.linspace(-1,2,100)
y_proba= sigmoid(-0.5+thetha*x_new)
decision_boundary= x_new[y_proba>=0.5][0]

plt.figure(figsize=(10,5))
plt.plot(x[y==0],y[y==0], 'bs')
plt.plot(x[y==1],y[y==1], 'g^')
plt.plot([decision_boundary,decision_boundary],[-1,2],'k:',linewidth=2)

plt.plot(x_new, y_proba, 'g-',label='Clase 1')
plt.plot(x_new,1-y_proba,'b--',label='Clase 0')
plt.xlabel('x',fontsize=14)
plt.ylabel('Probabilidad',fontsize=14)
plt.legend(loc='center right',fontsize=14)
plt.axis([-1,2,-0.02,1.02])
plt.show()
```

# Petal Width

```python
x=iris.data[:,3:]
x_b=np.c_[np.ones((len(iris.target),1)),x]
x_b
```

```python
# detect iris virginica
y=(iris.target==2).astype(int).reshape((len(iris.target),1))
y
```

```python
eta=0.9
n_iterations=50000
m=len(y)
```

```python
thetha=np.random.randn(2,1)
thetha 
```

```python
for i in range(n_iterations):
   gradients = (1/m)*(x_b.T@(sigmoid(x_b@thetha)-y))
   thetha = thetha - eta*gradients
thetha
```

```python

x_new_c=np.linspace(0,3,1000).reshape(-1,1)
x_new = np.c_[np.ones((1000,1)),x_new_c]
y_proba=sigmoid(thetha.T @ x_new.T)

decision_boundary= x_new_c[y_proba.reshape(-1,1)>=0.5][0]

plt.figure(figsize=(8,4))
plt.plot(x[y==0],y[y==0], 'bs')
plt.plot(x[y==1],y[y==1], 'g^')
plt.plot([decision_boundary,decision_boundary],[-1,2],'k:',linewidth=2)


plt.plot(x_new_c, y_proba.T, 'g-',label='IRIS VIRGINICA')
plt.plot(x_new_c,1-y_proba.T,'b--',label='NOT IRIS VIRGINICA')

plt.xlabel('Petal width(cm)',fontsize=12)
plt.ylabel('Probabilidad',fontsize=12)
plt.legend(loc='center right',fontsize=12)
plt.axis([0,3,-0.02,1.02])
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(C=10**10)
model.fit(x,y.ravel())
```

```python

x_new=np.linspace(0,3,1000).reshape(-1,1)

y_proba=model.predict_proba(x_new)

decision_boundary= x_new[y_proba[:,1]>=0.5][0]

plt.figure(figsize=(8,4))
plt.plot(x[y==0],y[y==0], 'bs')
plt.plot(x[y==1],y[y==1], 'g^')
plt.plot([decision_boundary,decision_boundary],[-1,2],'k:',linewidth=2)


plt.plot(x_new, y_proba[:,1], 'g-',label='IRIS VIRGINICA')
plt.plot(x_new,1-y_proba[:,1],'b--',label='NOT IRIS VIRGINICA')

plt.xlabel('Petal width(cm)',fontsize=12)
plt.ylabel('Probabilidad',fontsize=12)
plt.legend(loc='center right',fontsize=12)
plt.axis([0,3,-0.02,1.02])
plt.show()
```

```python
x=iris.data[:,(2,3)]
y = (iris.target==2).astype(int)
```

```python
model = LogisticRegression(C=10**10)
model.fit(x,y)
```

```python

left_right = np.array([0, 7])

try:
	intercept = model.intercept_[0]
	coef = model.coef_[0]
except AttributeError:

	model = LogisticRegression(C=10**10)
	model.fit(x, y)
	intercept = model.intercept_[0]
	coef = model.coef_[0]


boundary = -(intercept + coef[0] * left_right) / coef[1]

plt.figure(figsize=(8,4))
plt.plot(x[y==0, 0], x[y==0, 1], 'bs', label='Not Iris-Virginica')
plt.plot(x[y==1, 0], x[y==1, 1], 'g^', label='Iris-Virginica')

plt.plot(left_right, boundary, 'k:', linewidth=2)

plt.xlabel('Petal length (cm)', fontsize=12)
plt.ylabel('Petal width (cm)', fontsize=12)
plt.legend(loc='center left', fontsize=12)
plt.axis([0, 7, 0, 3.5])
plt.show()
```
